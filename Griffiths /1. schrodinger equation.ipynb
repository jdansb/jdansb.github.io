## Schrödinger equation

The Schrödinger equation plays a role in quantum mechanics analogous to that of Newton's second law in classical mechanics:

\begin{equation}
i\hbar\frac{\partial\Psi}{\partial t}=-\frac{\hbar^{2}}{2m}\frac{\partial^{2}\Psi}{\partial x^{2}}+V\Psi
\end{equation}

Where $\hbar$ is Planck's constant divided by $2\pi$. That is, if we have a particle of mass $m$ moving along the $x$ axis and we want to obtain information about the particle at any instant $t$, from an appropriate initial condition (typically $\Psi\left(x,0\right)$), the equation before determines the wave function $\Psi\left(x,t\right)$ for all future times, analogous to how Newton's second law determines the position $x\left(t\right)$ for all future instants.

But note that unlike position, wave function is not something we can measure directly. We will use it to calculate values that can be measured. For example, based on the wave function, Born's statistical interpretation provides us with the probability of finding the particle between points $a$ and $b$ at time $t$:

\begin{equation}
P=\int_{a}^{b}\left|\Psi\left(x,t\right)\right|^{2}dx
\end{equation}
