# Medida do trabalho contido

Apesar do foco do meu trabalho envolver o uso de modelos microscópicos para investigar sistemas complexos, a econofísica não se resume a isso. Entre alguns dos trabalhos considerados seminais na área, destacam-se os livros de Farjoun e Machover: *Laws of Chaos: A Probabilistic Approach to Political Economy* e *How Labor Powers the Global Economy*. A principal ideia nesses trabalhos é que a economia deve receber um tratamento estatístico.

O texto deste capítulo é retirado do apêndice do livro *How Labor Powers the Global Economy*, onde buscamos desenvolver ferramentas matemáticas para analisar o trabalho contido nas mercadorias. Logo, aqui não estamos apresentando um modelo ou lei, apenas um formalismo matemático para fazer análises.


Para começar, precisamos definir algumas propriedades do trabalho contido (tradução minha de *Labor Content*):

* O trabalho contido em uma coleção de produtos é igual ao total de horas de trabalho diretamente investidas na sua produção, mais a fração proporcional do trabalho contido nos produtos envolvidos no seu processo de produção.

* O trabalho contido na produção líquida de uma economia como um todo, durante um dado período, é igual ao total do número de horas trabalhadas durante o período.

Evidentemente, a segunda propriedade é uma conclusão da primeira, uma vez que, ao considerarmos toda a produção líquida da sociedade, necessariamente consideramos também toda a produção que foi utilizada no processo de produção. Desta forma, estamos levando em conta a produção como um todo da sociedade.


Vamos considerar uma economia com $m$ unidades de produção que conseguem,
durante um certo período de tempo (um ano por exemplo) produzir uma coleção
de produtos $B$. Uma parte $B_{I}$ destes produtos
será consumida durante o próprio processo de produção, resultando
assim em um produto líquido $B_{L}=B-B_{I}$.

Do produto líquido, uma parcela $B_{T}$ é destinadas aos trabalhadores - se considerarmos que a outra parte do produto líquido $B_{C}$ será
destinado aos capitalistas, logo $B_{L}=B_{T}+B_{C}$. O trabalho $L_{T}$ contido em $B_{T}$ será, da mesma forma uma parte do trabalho total $L$ contido no produto líquido  - De modo análogo podemos supor que o restante  $L_{C}$ do trabalho contido no produto líquido é o trabalho que está contido nos produtos destinados aos capitalistas de forma que $L=L_{T}+L_{C}$. Vamos chamar isso de partição do trabalho, e será denotado por:

\begin{equation}
\lambda=\frac{L_{T}}{L}
\end{equation}

Em outras palavras, apenas a fração $\lambda$ do trabalho contido no
produto líquido esta contida no produto destinado aos trabalhadores, o resto é apropriado por outros setores da sociedade. Vamos agora classificar
todos os produtos em $d$ distintos tipos diferentes. Então podemos
construir um vetor $\boldsymbol{b}=\left(b_{1},\dots,b_{d}\right)$
que registra a quantidade de cada tipo de produto que compõe o produto
líquido. Se definimos um vetor $\boldsymbol{c}=\left(c_{1},\dots,c_{d}\right)$
que descreve o trabalho contido por unidade de cada tipo de produto,
podemos calcular então o trabalho total contido no produto líquido
fazendo:

\begin{equation}
L=\boldsymbol{c}\cdot\boldsymbol{b}
\end{equation}

Se construímos um vetor $\boldsymbol{w}$ que representa a quantidade
de produtos líquidos distribuído para a força de trabalho dividido pelo trabalho
total empregado pela sociedade, temos $\frac{\boldsymbol{b}_{T}}{L}=\boldsymbol{w}$,
onde $\boldsymbol{b}_{T}$ é o vetor que representa a parte do produto
líquido que foi destinado aos trabalhadores. Evidentemente o trabalho
contido no produto líquido destinado aos trabalhadores será:

\begin{equation}
L_{T}=\boldsymbol{c}\cdot\boldsymbol{b}_{T}=\boldsymbol{c}\cdot\boldsymbol{w}L
\end{equation}

Lembrando que $\boldsymbol{c}$ não depende da quantidade de em uma dada coleção de mercadorias, então:

\begin{equation}
\lambda=\frac{L_{T}}{L}=\frac{\boldsymbol{c}\cdot\boldsymbol{w}L}{L}=\boldsymbol{c}\boldsymbol{w}
\end{equation}

Agora vamos conectar os produtos distribuídos para a força de trabalhos
com os produtos produzidos ao longo das $m$ unidades distintas de
produção. Vamos considerar que a firma $i$ emprega $x_{i}$ horas
de trabalho diretamente em um dado ano e $\boldsymbol{a}_{i}$ é um
vetor que especifica a quantidade de cada tipo de produto usada por
hora de trabalho diretamente empregado pela firma. Então os produtos
consumidos por uma firma ao longo do ano é $\boldsymbol{a}_{i}x_{i}$.
Por exemplo, se temos apenas dois produtos, teríamos para uma
firma $i$:

\begin{equation}
\boldsymbol{a}_{i}=\left(a_{1i},a_{2i}\right),\quad\boldsymbol{b}_{I_{i}}=x_{i}\boldsymbol{a}_{i}=\left(x_{i}a_{i1},x_{i}a_{i2}\right)
\end{equation}

Então podemos construir uma matriz $A$ onde cada coluna $j$ é um
vetor coluna $\boldsymbol{a}_{i}$ indicando consumo de produtos pela
firma $i$ por hora de trabalho empregada. Isto é, a linha $i$ indica o consumo do produto $i$ por cada empresa, e a coluna $j$ indica o consumo de cada produto
pela empresa $j$. Se temos por exemplo duas empresas, com dois tipos
de produtos, então:

\begin{equation}
\boldsymbol{A}=\left[\begin{array}{cc}
\boldsymbol{a}_{1} & \boldsymbol{a}_{2}\end{array}\right]=\left[\begin{array}{cc}
a_{11} & a_{12}\\
a_{21} & a_{22}
\end{array}\right]
\end{equation}

Logo, podemos obter o vetor $\boldsymbol{b}_{I}$ representando a
cesta de produtos consumidos durante o próprio processo de produção
ao longo de um ano utilizando uma matriz com os componentes $a_{ij}$:

\begin{equation}
\boldsymbol{A}\boldsymbol{x} & =\left[\begin{array}{cc}
a_{11} & a_{12}\\
a_{21} & a_{22}
\end{array}\right]\left[\begin{array}{c}
x_{1}\\
x_{2}
\end{array}\right]\\
 & =\left[a_{11}x_{1}+a_{12}x_{2},a_{21}x_{1}+a_{22}x_{2}\right]\\
 & =\left[a_{11}x_{1},a_{21}x_{1}\right]+\left[a_{12}x_{2},a_{22}x_{2}\right]\\
 & =\left[a_{11},a_{21}\right]x_{1}+\left[a_{12},a_{22}\right]x_{2}\\
 & =\boldsymbol{a}_{1}x_{1}+\boldsymbol{a}_{2}x_{2}\\
 & =\boldsymbol{b}_{I}
\end{equation}

Evidentemente este cálculo é facilmente generalizado para qualquer
quantidade $d$ de tipos de produtos e $m$ de firmas, onde $\boldsymbol{x}$
é um vetor coluna de dimensão $m$, e $A$ uma matriz de dimensão
$d\times m$:

\begin{equation}
\boldsymbol{A}\boldsymbol{x}=\left[\begin{array}{ccc}
a_{11} & \dots & a_{1m}\\
\vdots & \ddots & \vdots\\
a_{d1} & \dots & a_{dm}
\end{array}\right]\left[\begin{array}{c}
x_{1}\\
\vdots\\
x_{m}
\end{array}\right]
\end{equation}

Se $\boldsymbol{q}$ é o vetor que denota a produção bruta em
um ano, então é fácil de entender que:

\begin{equation}
\boldsymbol{b}=\boldsymbol{q}-\boldsymbol{b}_{I}\rightarrow\boldsymbol{q}=\boldsymbol{b}-\boldsymbol{A}\boldsymbol{x}
\end{equation}

Denotando o vetor de produtos produzidos por uma firma $i$
ao longo do ano como $\boldsymbol{q}_{i}=\left[q_{i1},\dots q_{id}\right]$,
então definindo que uma firma $i$, produz $p_{ij}q_{j}$ unidades
do tipo $j$, isto é, $q_{ij}=p_{ij}q_{j}$, onde $p_{ij}$ é a proporção
no qual a firma $i$ contribui no produto total da produção do produto
$j$ em toda economia, que é dada pelo componente $q_{j}$ do vetor
$\boldsymbol{q}$. Por exemplo, se uma firma $i$ produz 100\% do
produto $j$, simplesmente $p_{ij}=1$ e $q_{ji}=q_{j}$ e para toda
outra empresa diferente de $i$ temos $q_{ji}=p_{ij}=0$.

 Uma firma $i$ emprega direatmente $x_{i}$ unidades de trabalho para produzir
$d_{i}$ tipos diferentes de produtos. Então a quantidade
$l_{ij}=x_{i}/q_{ij}$ que vai nos dar a quantidade de horas de trabalho
necessárias serem empregadas diretamente pela firma $i$ produzir
uma unidade de produto do tipo $j$ . Ainda podemos reescrever $l_{ij}=x_{i}/\left(p_{ij}q_{j}\right)$. Podemos reescrever como:

\begin{equation}
x_{i}=l_{ij}p_{ij}q_{j}.
\end{equation}

Então se somarmos sobre todos os produtos $j$ que a firma $i$ produz
teríamos $\sum_{j}l_{ij}p_{ij}q_{j}=d_{i}x_{i}$, uma vez
que a firma $i$ produz $d_{i}$ produtos. Então podemos escrever
também:

\begin{equation}
x_{i}=\frac{1}{d_{i}}\sum_{j}l_{ij}p_{ij}q_{j}
\end{equation}

Vamos utilizar tudo isto para construir uma matriz $\boldsymbol{L}$
com elementos $\frac{l_{ij}p_{ij}}{d_{i}}$:

\begin{equation}
\boldsymbol{x}=\left[\begin{array}{c}
x_{1}\\
\vdots\\
x_{m}
\end{array}\right]=\left[\begin{array}{c}
\sum_{j}\frac{l_{1j}p_{1j}q_{j}}{d_{1}}\\
\vdots\\
\frac{\sum_{j}l_{mj}p_{mj}q_{j}}{d_{m}}
\end{array}\right]=\left[\begin{array}{ccc}
\frac{l_{11}p_{11}}{d_{1}} & \dots & \frac{l_{1d}p_{1d}}{d_{1}}\\
\vdots & \ddots & \vdots\\
\frac{l_{m1}p_{m1}}{d_{m}} & \dots & \frac{l_{md}p_{md}}{d_{m}}
\end{array}\right]\left[\begin{array}{c}
q_{1}\\
\vdots\\
q_{d}
\end{array}\right]=\boldsymbol{L}\boldsymbol{q}
\end{equation}

Então o par de matrizes $\left(\boldsymbol{A},\boldsymbol{L}\right)$
descreve as condições técnicas de produção. Podemos combinar os resultados:

\begin{equation}
\boldsymbol{b} & =\boldsymbol{q}-\boldsymbol{A}\boldsymbol{x}\\
 & =\boldsymbol{q}-\boldsymbol{A}\boldsymbol{L}\boldsymbol{q}\\
 & =\left(\boldsymbol{I}-\boldsymbol{A}\boldsymbol{L}\right)\boldsymbol{q}
\end{equation}

Então para que a gente tenha uma produção líquida maior que zero,
dependemos de $\boldsymbol{AL}$[^1]. Podemos reescrever tudo em uma única expressão:

\begin{equation}
\left[\begin{array}{c}
\boldsymbol{b}\\
\boldsymbol{0}
\end{array}\right]=\left[\begin{array}{cc}
\boldsymbol{I}_{d} & -\boldsymbol{A}\\
-\boldsymbol{L} & \boldsymbol{I}_{m}
\end{array}\right]\left[\begin{array}{c}
\boldsymbol{q}\\
\boldsymbol{x}
\end{array}\right]
\end{equation}

Onde estamos indicando por exemplo que se $\boldsymbol{u}=\left(u_{1},u_{2}\right)$
e $\boldsymbol{v}=\left(v_{1},v_{2}\right)$ então em uma notação sucinta
escrevemos $\left(\boldsymbol{u},\boldsymbol{v}\right)=\left(u_{1},u_{2},v_{1},v_{2}\right)$. Além disso, $\boldsymbol{0}$ é um vetor de valores nulos, $\boldsymbol{I}_{n}$
é uma matriz identidade de $n\times n$, isto é necessário pois para
uma multiplicação de matrizes $n\times m$ e $j\times k$ precisamos
que $m=j$, isto é, o número de colunas da primeira matriz precisa
ser igual o número de linhas da segunda. Então $\boldsymbol{I}_{d}$
é uma matriz $d\times d$ pra poder multiplicar o vetor $\boldsymbol{q}$
de tamanho $d$. Como $\boldsymbol{A}$ é uma matriz $d\times m$,
então a quantidade de colunas é $d+m=n$. Já $\boldsymbol{L}$é uma
matriz $m\times d$, de forma que a quantidade de linhas também é
$n=m+d$, o que complementa perfeitamente a última matriz identidade
$\boldsymbol{I}_{m}$ que multiplica o vetor $\boldsymbol{x}$ de
dimensão $m$. De forma que o vetor coluna também tem dimensão $n=m+d$.



Se aplicamos a matriz inversa:

\begin{equation}
\left[\begin{array}{cc}
\boldsymbol{I}_{d} & -\boldsymbol{A}\\
-\boldsymbol{L} & \boldsymbol{I}_{m}
\end{array}\right]^{-1} \left[\begin{array}{c}
\boldsymbol{b}\\
\boldsymbol{0}
\end{array}\right] & =\left[\begin{array}{c}
\boldsymbol{q}\\
\boldsymbol{x}
\end{array}\right]\\
\end{equation}



Agora podemos relacionar a quantidade de produtos atribuída a força de trabalho aos produtos que ela produz. Se $L=\sum_{i}x_{i}$ é a
quantidade total de horas de trabalho empregadas pelas firmas diretamente ao longo do ano, então se escrevermos um vetor de tamanho $m$ onde todos elementos
são $1$ como $\boldsymbol{1}$ então $\boldsymbol{1}\cdot\boldsymbol{x}=\sum_{i}x_{i}=L$.
Logo

\begin{equation}
\boldsymbol{b}_{T} & =\boldsymbol{w}L\\
 & =\boldsymbol{w}\left(\boldsymbol{1}\boldsymbol{x}\right)\\
 & =\boldsymbol{w}\left[\begin{array}{cc}
\boldsymbol{0} & \boldsymbol{1}\end{array}\right]\left[\begin{array}{c}
\boldsymbol{q}\\
\boldsymbol{x}\end{array}\right]\\
 & =\boldsymbol{w}\left[\begin{array}{cc}
\boldsymbol{0} & \boldsymbol{1}\end{array}\right]\left[\begin{array}{cc}
\boldsymbol{I}_{d} & -\boldsymbol{A}\\
-\boldsymbol{L} & \boldsymbol{I}_{m}
\end{array}\right]^{-1}\left[\begin{array}{c}
\boldsymbol{b}\\
\boldsymbol{0}
\end{array}\right]
\end{equation}

Uma matriz inversa $\boldsymbol{H}^{-1}$ é aquela no qual $\boldsymbol{H}^{-1}\boldsymbol{H}=\boldsymbol{I}$.
Escrevendo a

\begin{equation}
\boldsymbol{H}=\left[\begin{array}{cc}
\boldsymbol{I}_{d} & -\boldsymbol{A}\\
-\boldsymbol{L} & \boldsymbol{I}_{m}
\end{array}\right],\quad\boldsymbol{H}^{-1}= & \left[\begin{array}{cc}
\alpha_{11} & \alpha_{12}\\
\alpha_{21} & \alpha_{22}
\end{array}\right]
\end{equation}

Queremos descobrir os componentes $\alpha_{ij}$. Por hora vamos deixar
com esta notação.

\begin{equation}
\boldsymbol{b}_{T} & =\boldsymbol{w}\left[\begin{array}{cc}
\boldsymbol{0} & \boldsymbol{1}\end{array}\right]\left[\begin{array}{cc}
\alpha_{11} & \alpha_{12}\\
\alpha_{21} & \alpha_{22}
\end{array}\right]\left[\begin{array}{c}
\boldsymbol{b}\\
\boldsymbol{0}
\end{array}\right]\\
 & =\boldsymbol{w}\left[\begin{array}{cc}
\boldsymbol{1}\alpha_{21} & \boldsymbol{1}\alpha_{22}\end{array}\right]\left[\begin{array}{c}
\boldsymbol{b}\\
\boldsymbol{0}
\end{array}\right]\\
 & =\boldsymbol{w}\boldsymbol{1}\alpha_{21}\boldsymbol{b}
\end{equation}

Definindo $\boldsymbol{W}=\boldsymbol{w}\boldsymbol{1}\alpha_{21}$
então temos uma relação entre a produção líquida destinada aos trabalhadores
e a produção líquida no geral:

\begin{equation}
\boldsymbol{b}_{T}=W\boldsymbol{b}
\end{equation}


Então queremos o componente $\alpha_{21}$ da matriz inversa. Vamos
calcular a inversa explicitamente agora- lembrando que nesta notação, cada elemento da matriz $\boldsymbol{H}$ é outra matriz. Por isso os elementos do lado direito da igualdade também são matrizes (nula e identidade). E a matriz identidade à direita no total tem que ser necessariamente $n\times n$ mas temos uma certa flexibilidade pra escolher as dimensões das  sub-matrizes identidade. Definindo elas com ordens $n_{1}\times n_{1}$ e $n_{2}\times n_{2}$ desde que $n=n_{1}+n_{2}$,
e as matrizes nulas sejam o necessário para manter toda matriz quadrada.
Então fizemos a escolha de $n_{1}=d$ e $n_{2}=m$. Dessa forma as
matrizes nulas precisam ter as dimensões de $\boldsymbol{A}$ e $\boldsymbol{L}$:


\begin{equation}
\left[\begin{array}{cc}
\alpha_{11} & \alpha_{12}\\
\alpha_{21} & \alpha_{22}
\end{array}\right]\left[\begin{array}{cc}
\boldsymbol{I}_{d} & -\boldsymbol{A}\\
-\boldsymbol{L} & \boldsymbol{I}_{m}
\end{array}\right] & =\left[\begin{array}{cc}
\boldsymbol{I}_{d} & \boldsymbol{0}_{A}\\
\boldsymbol{0}_{L} & \boldsymbol{I}_{m}
\end{array}\right]\\
\left[\begin{array}{cc}
\alpha_{11}\boldsymbol{I}_{d}-\alpha_{12}\boldsymbol{L} & -\alpha_{11}\boldsymbol{A}+\alpha_{12}\boldsymbol{I}_{m}\\
\alpha_{21}\boldsymbol{I}_{d}-\alpha_{22}\boldsymbol{L} & -\alpha_{21}\boldsymbol{A}+\alpha_{22}\boldsymbol{I}_{m}
\end{array}\right] & =\left[\begin{array}{cc}
\boldsymbol{I}_{d} & \boldsymbol{0}_{A}\\
\boldsymbol{0}_{L} & \boldsymbol{I}_{m}
\end{array}\right]
\end{equation}


Podemos notar que por definição os elementos $\alpha_{i1}$ precisam
ter $d$ colunas para multiplicar $\boldsymbol{I}_{d}$ e $\boldsymbol{A}$,
análogo a $\alpha_{i2}$ que precisa ter $m$ colunas para multiplicar
$\boldsymbol{L}$ e $\boldsymbol{I}_{m}$. De forma que:

\begin{equation}
\left[\begin{array}{cc}
\alpha_{11}-\alpha_{12}\boldsymbol{L} & -\alpha_{11}\boldsymbol{A}+\alpha_{12}\\
\alpha_{21}-\alpha_{22}\boldsymbol{L} & -\alpha_{21}\boldsymbol{A}+\alpha_{22}
\end{array}\right]=\left[\begin{array}{cc}
\boldsymbol{I}_{d} & \boldsymbol{0}_{A}\\
\boldsymbol{0}_{L} & \boldsymbol{I}_{m}
\end{array}\right]
\end{equation}

Ficamos então com 2 sistemas:

\begin{equation}
\alpha_{11}-\alpha_{12}\boldsymbol{L} & =\boldsymbol{I}_{d} & \alpha_{21}-\alpha_{22}\boldsymbol{L} & =\boldsymbol{0}_{L}\\
\alpha_{11}\boldsymbol{A}+\alpha_{12} & =\boldsymbol{0}_{A} & -\alpha_{21}\boldsymbol{A}+\alpha_{22} & =\boldsymbol{I}_{m}
\end{equation}

Do segundo sistema, e da segunda linha temos que:

\begin{equation}
\alpha_{22} & =\boldsymbol{I}_{m}+\alpha_{21}\boldsymbol{A}
\end{equation}

Substituindo na primeira linha:

\begin{equation}
\alpha_{21}-\alpha_{22}\boldsymbol{L} & =\boldsymbol{0}_{L}\\
\alpha_{21}-\left(\boldsymbol{I}_{m}+\alpha_{21}\boldsymbol{A}\right)\boldsymbol{L} & =\boldsymbol{0}_{L}\\
\alpha_{21}-\boldsymbol{I}_{m}\boldsymbol{L}-\alpha_{21}\boldsymbol{A}\boldsymbol{L} & =\boldsymbol{0}_{L}\\
\alpha_{21}-\alpha_{21}\boldsymbol{A}\boldsymbol{L} & =\boldsymbol{0}_{L}+\boldsymbol{L}\\
\alpha_{21}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right) & =\boldsymbol{L}
\end{equation}

Vale destacar como $\boldsymbol{AL}$ é uma matriz de ordem $d\times m$
multiplicando uma matriz de ordem $m\times d$, resultando em uma
matriz de ordem $d\times d$. Ou seja, multiplicando toda igualdade pela sua inversa
à direita, temos:

\begin{equation}
\alpha_{21}=\boldsymbol{L}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right)^{-1}
\end{equation}

E então:

\begin{equation}
\boldsymbol{W} & =\boldsymbol{w}\boldsymbol{1}\alpha_{12}\\
 & =\boldsymbol{w}\boldsymbol{1}\boldsymbol{L}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right)^{-1}
\end{equation}

O livro chama $\boldsymbol{W}$ de <<matriz de alocação de trabalhadores>>.
Lembrando da equação $\boldsymbol{b}_{T}=\boldsymbol{W}\boldsymbol{b}^{T}$,
ela não diz exatamente como os trabalhadores são alocados, mas da
produção líquida $\boldsymbol{b}$, qual a parte da produção é alocada
aos trabalhadores $\boldsymbol{b}_{T}$. Se podemos escrever $\boldsymbol{b}_{T}=\boldsymbol{W}\boldsymbol{b}$
então:

\begin{equation}
\lambda & =\frac{L_{T}}{L}\\
 & =\frac{\boldsymbol{c\cdot Wb}}{\boldsymbol{c}\cdot\boldsymbol{b}}\\
\lambda\boldsymbol{c}\cdot\boldsymbol{b} & =\boldsymbol{c\cdot Wb}\\
\lambda\boldsymbol{c}\cdot\boldsymbol{b}-\boldsymbol{c\cdot Wb} & =0
\end{equation}

Lembrando que $\lambda=\boldsymbol{c}\cdot\boldsymbol{w}$, isso nos convida a considerar que $\lambda$ é independente de $\boldsymbol{b}$, ou seja, a partição
do trabalho depende de alguns aspectos técnicos do modo de produção
dado por $\boldsymbol{c}$ e não da produção líquida $\boldsymbol{b}$.
Então

\begin{equation}
\lambda\boldsymbol{c}\cdot\boldsymbol{b}-\boldsymbol{c\cdot Wb} & =0\\
\left(\lambda\boldsymbol{c}-\boldsymbol{cW}\right) & =0\\
\lambda\boldsymbol{c} & =\boldsymbol{cW}\\
\left(\lambda\boldsymbol{c}\right)^{T} & =\left(\boldsymbol{cW}\right)^{T}\\
\boldsymbol{c}^{T}\lambda & =\boldsymbol{W}^{T}\boldsymbol{c}^{T}\\
0 & =\left(\boldsymbol{W}^{T}-\lambda\boldsymbol{I}\right)\boldsymbol{c}^{T}
\end{equation}

Então $\lambda$ é o auto valor dado por $\det\left(\boldsymbol{W}^{T}-\lambda\boldsymbol{I}\right)=0$
e como $\det\left(\boldsymbol{W}^{T}-\lambda\boldsymbol{I}\right)=\det\left(\boldsymbol{W}-\lambda\boldsymbol{I}\right)$,
queremos simplesmente os autovalores de $\boldsymbol{W}$. Sendo $\boldsymbol{W}=\boldsymbol{w}\boldsymbol{1}\boldsymbol{L}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right)^{-1}$,
então:

\begin{equation}
\det\left(\boldsymbol{w}\boldsymbol{1}\boldsymbol{L}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right)^{-1}-\lambda\boldsymbol{I}\right)=0
\end{equation}

Usando o [lema do determinante de matriz](https://pt.wikipedia.org/wiki/Lema_do_determinante_de_matriz):

\begin{equation}
\det\left(A+uv^{T}\right) & =\left(1+v^{T}A^{-1}u\right)\det\left(A\right)
\end{equation}

Então sendo $A=-\lambda\boldsymbol{I}$ , $u=\boldsymbol{w}$ e $v^{T}=\boldsymbol{1}\boldsymbol{L}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right)^{-1}$,
então:

\begin{equation}
\det\left(\boldsymbol{W}^{T}-\lambda\boldsymbol{I}\right) & =\left(1+\left[\boldsymbol{1}\boldsymbol{L}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right)^{-1}\right]\left(-\lambda\boldsymbol{I}\right)^{-1}\boldsymbol{w}\right)\det\left(-\lambda\boldsymbol{I}\right)\\
 & =\left(1-\boldsymbol{1}\boldsymbol{L}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right)^{-1}\boldsymbol{I}\boldsymbol{w}\lambda^{-1}\right)\det\left(-\lambda\boldsymbol{I}\right)\\
 & =\left(1-\boldsymbol{1}\boldsymbol{L}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right)^{-1}\boldsymbol{w}\lambda^{-1}\right)\left(-\lambda\right)^{d}\det\left(\boldsymbol{I}\right)\\
 & =\left(1-\boldsymbol{1}\boldsymbol{L}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right)^{-1}\boldsymbol{w}\lambda^{-1}\right)\left(-\lambda\right)^{d}\\
 & =\left(1-\boldsymbol{1}\boldsymbol{L}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right)^{-1}\boldsymbol{w}\lambda^{-1}\right)\lambda^{d}\left(-1\right)^{d}
\end{equation}

Então:

\begin{equation}
0= & \det\left(\boldsymbol{W}^{T}-\lambda\boldsymbol{I}\right)\\
0= & \left(1-\boldsymbol{1}\boldsymbol{L}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right)^{-1}\boldsymbol{w}\lambda^{-1}\right)\lambda^{d}\left(-1\right)^{d}\\
1= & \frac{\boldsymbol{1}\boldsymbol{L}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right)^{-1}\boldsymbol{w}}{\lambda}\\
\lambda= & \boldsymbol{1}\boldsymbol{L}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right)^{-1}\boldsymbol{w}
\end{equation}

Retomando então:

\begin{equation}
\lambda\boldsymbol{c} & =\boldsymbol{cW}\\
\boldsymbol{c} & =\frac{1}{\lambda}\boldsymbol{cW}\\
 & =\frac{\boldsymbol{c}\boldsymbol{w}}{\lambda}\boldsymbol{1}\boldsymbol{L}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right)^{-1}
\end{equation}

E como vimos que $\lambda=\boldsymbol{c}\boldsymbol{w}$

\begin{equation}
\boldsymbol{c} & =\frac{\boldsymbol{c}\boldsymbol{w}}{\boldsymbol{c}\boldsymbol{w}}\boldsymbol{1}\boldsymbol{L}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right)^{-1}\\
 & =\boldsymbol{1}\boldsymbol{L}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right)^{-1}
\end{equation}


O trabalho contido em uma coleção de produtos $\boldsymbol{b}$ qualquer pode
ser escrito então como:

\begin{equation}
L\equiv\boldsymbol{c}\boldsymbol{b}=\boldsymbol{1}\boldsymbol{L}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right)^{-1}\boldsymbol{b}
\end{equation}

Pode-se observar que a medida do trabalho contido $L$ calcula a média
do trabalho gasto ao longo de todas unidades de produção e acrescenta
todos os trabalhos intermediários necessários a cada estágio de produção.
Se usarmos a expansão em série de uma matriz em $L=\boldsymbol{1}\boldsymbol{L}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right)^{-1}\boldsymbol{b}$,
obtemos:

\begin{equation}
L=\left(\boldsymbol{c}_{0}+\boldsymbol{c}_{1}+\dots\right)\boldsymbol{b}
\end{equation}

Onde $\boldsymbol{c}_{n}=\boldsymbol{1}\boldsymbol{L}\left(\boldsymbol{A}\boldsymbol{L}\right)^{n}$
determina o trabalho necessário no enésimo estágio de produção (contando
do fim para trás). Dessa forma $L$ considera todo o trabalho necessário
para produzir uma coleção de produtos.


[^1]: Dessa relação podemos obter que é preciso que o maior autovalor de
$\boldsymbol{AL}$ seja menor que $1$ para que exista reprodução
econômica. No momento estou reproduzindo a nota de rodapé do livro, eu preciso

## Exemplo

Vamos imaginar uma sociedade com duas unidades de produção (Um e
Dois) que durante um ano produzem Aço e Banana. A produção total no
ano é de $30$ unidades de aço e $40$ unidades de banana, ou seja
$\boldsymbol{q}=\left(30,40\right)$, disto metade é produção líquida
$\boldsymbol{b}=\left(15,20\right)$ e aproximadamente metade da produção
líquida é destinada aos trabalhadores $\boldsymbol{b}_{T}=\left(5,10\right)$.

As 2 firmas juntas empregam  $L=250$ horas de trabalho anual, vamos
considerar que a Firma Um emprega 150h e a Firma Dois 100h, ou seja
$\boldsymbol{x}=\left(150,100\right)$. A produção que é consumida
internamente, conforme definimos, é metade da produção total, ou seja
$\boldsymbol{b}_{I}=\boldsymbol{b}=\left(15,20\right)$. Considerando
que a empresa Um consome 15 aço e 10 bananas então $\boldsymbol{b}_{I_{1}}=\left(15,10\right)$.
A Firma Dois consome o restante: 0 unidades de aço e 10 unidades de
banana, ou seja $\boldsymbol{b}_{I_{2}}=\left(0,10\right)$. Podemos também calcular o vetor $\boldsymbol{w}=\frac{\boldsymbol{b}_{T}}{L}=\left(\frac{5}{250},\frac{10}{250}\right)=\left(0.02,0.04\right)$,
isto é, para cada unidade de trabalho que a classe trabalhadora emprega
diretamente, ela ganha $0.02$ unidades de aço e $0.04$ unidades
de banana.

 Como
$\boldsymbol{b}_{I_{i}}=\left(x_{i}a_{1i},\dots,x_{i}a_{di}\right)$
logo $\frac{b_{I_{ji}}}{x_{i}}=a_{ji}$, ou seja para o consumo de
aço da Firma Um temos: $a_{11}=\frac{15}{150}=0.1$, isso significa
que para cada 1h de trabalho empregado diretamente pela Firma Um,
ela precisa consumir $0.1$ unidade de aço. Podemos calcular de forma
equivalente:

\begin{equation}
a_{21}=\frac{10}{150}=\frac{1}{15},\quad a_{12}=\frac{0}{100}=0,\quad,a_{22}=\frac{10}{100}=0.1
\end{equation}

Vemos que de forma semelhante, pra cada uma hora de trabalho direto
empregado pela Firma Dois, ela precisa consumir $0.1$ unidade de
banana. Temos então $\boldsymbol{a}_{1}=\left(0.1,1/15\right)$ e
$\boldsymbol{a}_{2}=\left(0,0.1\right)$. Podemos construir a matriz
$\boldsymbol{A}$:

\begin{equation}
\boldsymbol{A}=\left[\begin{array}{cc}
0.1 & 0\\
1/15 & 0.1
\end{array}\right]\rightarrow\boldsymbol{b}_{I}=\boldsymbol{A}\boldsymbol{x}=\left[\begin{array}{cc}
0.1 & 0\\
1/15 & 0.1
\end{array}\right]\left[\begin{array}{c}
150\\
100
\end{array}\right]=\left[\begin{array}{c}
15\\
20
\end{array}\right]
\end{equation}


Agora quanto a produção bruta, considerando que a Firma Um produz
todo o aço e metade das bananas temos $\boldsymbol{q}_{1}=\left(30,20\right)$,
logo a Firma Dois deve produzir o restante $\boldsymbol{q}_{2}=\left(0,20\right)$
sendo que evidentemente a soma de ambos deve dar a produção bruta total
$\boldsymbol{q}=\boldsymbol{q}_{1}+\boldsymbol{q}_{2}$ conforme definida
no começo. Outra forma de colocar isso é dizendo que a Firma Um é
responsável por produzir 100\% do aço ($p_{11}=1$ e $p_{21}=0$)
e quanto a banana cada firma produz metade ( $p_{12}=p_{22}=0.5$).
Sendo que a Firma Um produz dois produtos $d_{1}=2$ e a Firma Dois
apenas um $d_{2}=1$, podemos calcular os termos $l_{ij}$ por ambos
os caminhos por exemplo $l_{12}$(sobre a produção de banana da Firma
Um):

\begin{equation}
l_{12} & =\frac{x_{1}}{q_{12}}=\frac{150}{20}=7.50\\
l_{12} & =\frac{x_{1}}{p_{12}q_{2}}=\frac{150}{0.5\cdot40}=\frac{150}{20}=7.50
\end{equation}

Ou seja, a Firma Um precisa de $7.50$ horas pra produzir uma unidade
de banana. Para os outros produtos:

\begin{equation}
l_{11} & =\frac{x_{1}}{q_{11}}=\frac{150}{30}=5.00\\
l_{21} & =\frac{x_{2}}{q_{21}}=\frac{100}{0}\approx\lim_{x\rightarrow0}\frac{100}{0}=\infty\\
l_{22} & =\frac{x_{2}}{q_{22}}=\frac{100}{20}=5.00
\end{equation}

Aqui é interessante notar que $l_{21}$ eu usei uma aproximação que
nos indica que a Firma Dois precisa de infinitas horas pra produzir
aço, pois ela não produz. Vale notar que não estamos definindo diretamente
sobre como o futuro da produção bruta de cada firma se divide entre
entre consumo interno para a produção e e produção líquida, nem sobre
como os produtos consumidos pela firma se distribuem entre seus produtos
líquidos.

Agora somos capazes de construir a matriz $\boldsymbol{L}$:

\begin{equation}
\boldsymbol{L}=\left[\begin{array}{cc}
\frac{l_{11}p_{11}}{d_{1}} & \frac{l_{12}p_{12}}{d_{1}}\\
\frac{l_{21}p_{21}}{d_{2}} & \frac{l_{22}p_{22}}{d_{2}}
\end{array}\right]=\left[\begin{array}{cc}
\frac{5}{2} & \frac{7.5\cdot0.5}{2}\\
0 & 5\cdot0.5
\end{array}\right]=\left[\begin{array}{cc}
2.5 & 1.875\\
0 & 2.5
\end{array}\right]
\end{equation}

Podemos recuperar $\boldsymbol{x}$:

\begin{equation}
\boldsymbol{x}=\boldsymbol{L}\boldsymbol{q}=\left[\begin{array}{cc}
2.5 & 1.875\\
0 & 2.5
\end{array}\right]\left[\begin{array}{c}
30\\
40
\end{array}\right]=\left[\begin{array}{c}
150\\
100
\end{array}\right]
\end{equation}

Ou seja, podemos ter a produção líquida fazendo:

\begin{equation}
\boldsymbol{b} & =\left(\boldsymbol{I}-\boldsymbol{A}\boldsymbol{L}\right)\boldsymbol{q}\\
 & = \left(\left[\begin{array}{cc}
1 & 0\\
0 & 1
\end{array}\right]-\left[\begin{array}{cc}
0.1 & 0\\
1/15 & 0.1
\end{array}\right]\left[\begin{array}{cc}
2.5 & 1.875\\
0 & 2.5
\end{array}\right]\right)\left[\begin{array}{c}
30\\
40
\end{array}\right]\\
 & = \left(\left[\begin{array}{cc}
1 & 0\\
0 & 1
\end{array}\right]-\left[\begin{array}{cc}
0.25 & 0.1875\\
2.5/15 & 0.375
\end{array}\right]\right)\left[\begin{array}{c}
30\\
40
\end{array}\right]\\
 & =\left[\begin{array}{cc}
0.75 & -0.1875\\
-2.5/15 & 0.625
\end{array}\right]\left[\begin{array}{c}
30\\
40
\end{array}\right]\\
 & =\left[\begin{array}{c}
15\\
20
\end{array}\right]
\end{equation}

Por fim então, podemos calcular a matriz de alocação de trabalhadores:

\begin{equation}
\boldsymbol{W}=\boldsymbol{w}\boldsymbol{1}\boldsymbol{L}\left(\boldsymbol{I}_{d}-\boldsymbol{A}\boldsymbol{L}\right)^{-1}=\boldsymbol{w}\boldsymbol{1}\boldsymbol{L}\boldsymbol{M}^{-1}
\end{equation}

Como podemos ver $\boldsymbol{M}$ é a matriz que calculamos agora
em $\boldsymbol{b}=\boldsymbol{M}\boldsymbol{q}$. Então podemos calcular
a inversa fazendo:

\begin{equation}
\boldsymbol{M}^{-1}\boldsymbol{M} & =\boldsymbol{I}\\
\left[\begin{array}{cc}
a_{11} & a_{12}\\
a_{21} & a_{22}
\end{array}\right]\left[\begin{array}{cc}
0.75 & -0.1875\\
-2.5/15 & 0.625
\end{array}\right] & = \left[\begin{array}{cc}
1 & 0\\
0 & 1
\end{array}\right]\\
\left[\begin{array}{cc}
0.75a_{11}-a_{12}2.5/15 & -0.1875a_{11}+a_{12}0.625\\
0.75a_{21}-a_{22}2.5/15 & -0.1875a_{21}+a_{22}0.625
\end{array}\right] & = \left[\begin{array}{cc}
1 & 0\\
0 & 1
\end{array}\right]
\end{equation}

Temos então:

\begin{equation}
0.75a_{11}-a_{12}2.5/15 & =1 & 0.75a_{21}-a_{22}2.5/15 & =0\\
-0.1875a_{11}+a_{12}0.625 & =0 & -0.1875a_{21}+a_{22}0.625 & =1
\end{equation}

Dessa vez precisamos resolver os dois sistemas. Trabalhando com a
linha de baixo do sistema a esquerda e a de cima com o sistema à direita:

\begin{equation}
-a_{11}+a_{12}\frac{0.625}{0.1875} & =0 & a_{21}-a_{22}\frac{2.5}{15\cdot0.75} & =0\\
a_{11} & =a_{12}\frac{0.625}{0.1875} & a_{21} & =a_{22}\frac{2.5}{11.25}
\end{equation}

Substituindo na outra linha:

\begin{equation}
0.75a_{11}-a_{12}\frac{2.5}{15} & =1 & -0.1875a_{21}+a_{22}0.625 & =1\\
0.75\left(a_{12}\frac{0.625}{0.1875}\right)-a_{12}\frac{2.5}{15} & =1 & -0.1875\left(a_{22}\frac{2.5}{11.25}\right)+a_{22}0.625 & =1\\
a_{12}\left(\frac{0.46875}{0.1875}-\frac{2.5}{15}\right) & =1 & a_{22}\left(-\frac{0.46875}{11.25}+0.625\right) & =1\\
2.3333a_{12} & =1 & 0.58333a_{22} & =1\\
a_{12} & =2.3333 & a_{22} & =\frac{1}{0.58333}\\
a_{12} & =0.42857 & a_{22} & =1.71429
\end{equation}

Podemos calcular os elementos restantes:

\begin{equation}
a_{11} & =a_{12}\frac{0.625}{0.1875} & a_{21} & =a_{22}\frac{2.5}{11.25}\\
a_{11} & =0.42857\cdot\frac{0.625}{0.1875} & a_{21} & =1.71429\cdot\frac{2.5}{11.25}\\
a_{11} & =1.42856 & a_{21} & =0.38095
\end{equation}

Então:

\begin{equation}
\boldsymbol{M}^{-1}=\left[\begin{array}{cc}
a_{11} & a_{12}\\
a_{21} & a_{22}
\end{array}\right]=\left[\begin{array}{cc}
1.42857 & 0.42857\\
0.38095 & 1.71429
\end{array}\right]
\end{equation}

Agora se fazemos:

\begin{equation}
\boldsymbol{W} & =\boldsymbol{w}\boldsymbol{1}\boldsymbol{L}\boldsymbol{M}^{-1}\\
 & =\left[\begin{array}{c}
0.02\\
0.04
\end{array}\right]\left[\begin{array}{cc}
1 & 1\end{array}\right]\left[\begin{array}{cc}
2.5 & 1.875\\
0 & 2.5
\end{array}\right]\left[\begin{array}{cc}
1.42857 & 0.42857\\
0.38095 & 1.71429
\end{array}\right]\\
 & =\left[\begin{array}{c}
0.02\\
0.04
\end{array}\right]\left[\begin{array}{cc}
1 & 1\end{array}\right]\left[\begin{array}{cc}
4.28571 & 4.28572\\
0.95237 & 4.28573
\end{array}\right]\\
 & =\left[\begin{array}{c}
0.02\\
0.04
\end{array}\right]\left[\begin{array}{cc}
5.2308 & 8.57144\end{array}\right]
\end{equation}

Ou seja:

\begin{equation}
\boldsymbol{W}=\left[\begin{array}{cc}
0.10476 & 0.17143\\
0.20952 & 0.34286
\end{array}\right]
\end{equation}

Lembrando que $L=\boldsymbol{1}\boldsymbol{x}=\left[\begin{array}{cc}
1 & 1\end{array}\right]\left[\begin{array}{c}
150\\
100
\end{array}\right]=150+100=250$.Podemos conferir fazendo:

\begin{equation}
\boldsymbol{b}_{T}=W\boldsymbol{b}=\left[\begin{array}{cc}
0.10476 & 0.17143\\
0.20952 & 0.34286
\end{array}\right]\left[\begin{array}{c}
15\\
20
\end{array}\right]=\left[\begin{array}{c}
5\\
10
\end{array}\right]
\end{equation}

Acredito que podemos interpretar cada termo $W_{ij}$ como a fração
do produto líquido $i$ que é destinado aos trabalhadores para cada
unidade de produto líquido $j$. Isto é, por exemplo $W_{11}\approx0.1$
indica que para cada 10 unidades de aço de produto líquido produzida,
1 uma unidade de aço é destinada aos trabalhadores ($0.1\cdot15=1.5$),
e $W_{12}\approx0.17$ de modo análogo indica que para cada unidade
de banana de produto líquido, $0.17$ unidade de aços é destinada
aos trabalhadores ($0.17\cdot20\approx3.5$) resultando então que
para essa produção líquida $\boldsymbol{b}$, $5$ unidades de aço
são produzidas para os trabalhadores.

Podemos agora calcuar o vetor  $\boldsymbol{c}$:

\begin{equation}
\boldsymbol{c}=\boldsymbol{1}\boldsymbol{L}\boldsymbol{M}^{-1}=\left[\begin{array}{cc}
1 & 1\end{array}\right]\left[\begin{array}{cc}
2.50 & 1.86\\
0.00 & 2.50
\end{array}\right]\left[\begin{array}{cc}
1.43 & 0.43\\
0.38 & 1.70
\end{array}\right]=\left[\begin{array}{cc}
5.23 & 8.57\end{array}\right]
\end{equation}

Podemos recuperar então o total de horas de trabalhadas no nosso exemplo:

\begin{equation}
L=\left[\boldsymbol{1}\boldsymbol{L}\left(I-\boldsymbol{A}\boldsymbol{L}\right)^{-1}\right]\boldsymbol{b}=\left[\begin{array}{cc}
5.23 & 8.57\end{array}\right]\left[\begin{array}{c}
15\\
20
\end{array}\right]=250
\end{equation}

E com isso podemos calcular a partição do trabalho:

\begin{equation}
\lambda=\boldsymbol{1}\boldsymbol{L}\boldsymbol{M}^{-1}\boldsymbol{w}=\boldsymbol{c}\boldsymbol{w}=\left[\begin{array}{cc}
5.23 & 8.57\end{array}\right]\left[\begin{array}{c}
0.02\\
0.04
\end{array}\right]=0.4476
\end{equation}


Vamos interpertar estes resultados. Cada unidade de aço demanda em média $5.23$ horas de trabalho para ser produzida, e cada unidade de banana demanda $8.57$, então $\boldsymbol{c}=\left(5.23,8,57\right)$.

Então o trabalho contido na produção líquida é $L=\boldsymbol{c}\cdot\boldsymbol{b}=5.23\cdot15+8.57\cdot20\approx250$,
e o trabalho contido na produção destinada aos trabalhadores é $L_T=\boldsymbol{c}\cdot\boldsymbol{b_T}=5.23\cdot 5+8.57\cdot 10\approx 111.85$,
logo a partição do trabalho é dada por $\lambda=111.85/250 \approx 0.45$, aproxiadamente o mesmo resulado encontrado anteriormente.

Isso significa ques 45\% das horas empregados pelos trabalhadores diretamente no
sistema produtivo dessa sociedade estão contidas nos produtos
que são destinados a classe trabalhadora, o restante 55\% foi apropriado por
outra classe que não está diretamente envolvida no sistema de produção.
Essa aqui é uma divisão ainda por classe, é necessário ainda lembrar que
tipicamente a classe trabalhadora é muito mais numerosa que a classe capitalista.  Se considerarmos uma proporção de 10 trabalhadores para cada capitalista, significa que os 45\% do trabalho contido destinado aos trabalhadores é destinados para 10x mais
pessoas que os 55\% para os capitalistas.

Vamos calcular o trabalho
contido per capita, isto é, se temos $N$ capitalistas, e 55\%
é da produção é apropriado pelos capitalistas, em média cada capitalista
se apropria de $\frac{55}{N}\%$ da produção. Fazendo o mesmo para
os trabalhadores, mas considerando que tem $10N$ trabalhadores, então
cada trabalhador se apropria de $\frac{45}{10N}\%$ da produção. Agora
calculando a razão, isto é, quantas vezes mais o capitalista se apropria
da produção que o trabalhador temos:

\begin{equation}
\frac{\frac{55}{N}\%}{\frac{45}{10N}\%}=\frac{55}{N}\frac{10N}{45}=\frac{550}{45}\approx 12
\end{equation}

Então nete caso, um capitalista se apropria $12$ vezes mais da produção do que um
trabalhador. Se considerarmos que há 100 trabalhadores para cada capitalista, então este capitalista se aprorima de mais de $122$ vezes da produção do que comaprada a um trabalhador.  Considerado um capitalista que retira seus lucros simplesmente do fato de ter a posse dos meios de produção privada.
